## 4. Let's start Analysing the Dataset

```python
#Total Number of Sixes hit in the season
total_6=cwc[cwc['runs_off_bat']==6]['runs_off_bat'].count()
total_6
```

```python
#Total Number of fours hit in the season
total_4=cwc[cwc['runs_off_bat']==4]['runs_off_bat'].count()
total_4
```

```python
# Let's graph a pie chart to visualize the number of sixes and the number of fours hit in ICC CWC 2023
plt.figure(figsize=(4,2))
plt.pie([total_6,total_4],labels=["Sixes","Fours"],autopct="%1.1f%%")
plt.title('Number of Sixes and Fours hit in ICC CWC 2023')
```

```python
#Teams played the season
team=cwc['batting_team'].unique()
print("Team palyed in ICC World Cup 2023:-\n")
for i,team in enumerate(team,start=1):
    print(f"{i}:{team}")
```

```python
#Top 10 highest runs scorer of the season
highest_run=cwc.groupby('striker')['runs_off_bat'].sum().sort_values(ascending=False).head(10)
highest_run
```

```python
# Lets Plot the bar plot of the top 10 highest runs scorer of ICC CWC 2023 
plt.figure(figsize=(5,4))
plt.bar(highest_run.index,highest_run,color='skyblue',edgecolor='black')
plt.title('Top 10 highest runs scorer of ICC CWC 2023')
plt.xlabel('Batsman')
plt.ylabel('Runs')
plt.xticks(rotation=90);

```

```python
#Top 10 players who hit most number of sixes in the season
top_sixes=cwc[cwc['runs_off_bat']==6]
top_sixes_in_cwc=top_sixes.groupby('striker')['runs_off_bat'].count().sort_values(ascending=False).head(10)
top_sixes_in_cwc
```

```python
# Lets plot the line plot of the top 10 players who hit most number of sixes in ICC CWC 2023
plt.figure(figsize=(5,4))
plt.bar(top_sixes_in_cwc.index,top_sixes_in_cwc,color='skyblue',edgecolor='black')
plt.title("Top 10 players who hit most number of sixes in ICC CWC 2023")
plt.xlabel('Batsman')
plt.ylabel('Most sixes')
plt.xticks(rotation=90);
```

```python
#Top 10 batsmen who hit most number of sixes in sigle match
top_sixes=cwc[cwc['runs_off_bat']==6]
most_sixes=top_sixes.groupby(['match_id','striker'])['runs_off_bat'].count().sort_values(ascending=False).head(10)
most_sixes
```

```python
# Reseting the index of result1
most_sixes = most_sixes.reset_index()
most_sixes
```

```python
## Lets plot a bar plot of Top 10 batsmen who hit most number of sixes in sigle match
plt.figure(figsize=(5,4))
plt.bar(most_sixes['striker'], most_sixes['runs_off_bat'], color='skyblue', edgecolor='black')
plt.title('Top 10 Batsmen with Most Sixes in a Single Match')
plt.xlabel('Batsman')
plt.ylabel('Number of Sixes')
plt.xticks(rotation=90);
```

```python
#Most Sixes hit in a Venue
top_sixes=cwc[cwc['runs_off_bat']==6]
venue_sixes = top_sixes['venue'].value_counts().head(10)
venue_sixes
```

```python
# Lets plot the bar plot of the number of sixes hit in a venue
plt.figure(figsize=(4,4))
plt.bar(venue_sixes.index,venue_sixes,color='skyblue',edgecolor='black')
plt.title("Number of sixes hit in a venue")
plt.xlabel("Venue Name")
plt.ylabel("No. of sixes")
plt.xticks(rotation=90);
```

```python
# Top 10 players who hit most number of fours in the season

top_fours = cwc[cwc['runs_off_bat'] == 4]
top_batsman_4 = top_fours.groupby('striker')['runs_off_bat'].count().sort_values(ascending=False).head(10)
top_batsman_4
```

```python
# Lets plot the top 10 players who hit most number of foures in ICC CWC 2023
plt.figure(figsize=(5,4))
plt.bar(top_batsman_4.index,top_batsman_4,color='skyblue',edgecolor='black')
plt.title("Top 10 players who hit most number of foures in ICC CWC 2023")
plt.xlabel('Batsman')
plt.ylabel('No.of foures')
plt.xticks(rotation=90);
```

```python
#Top 10 batsmen who hit most number of fours in sigle match
top_fours = cwc[cwc['runs_off_bat'] == 4]
top_batsman=top_fours.groupby(['match_id','striker'])['runs_off_bat'].count().sort_values(ascending=False).head(10)
top_batsman
```

```python
# reset_index of result2
top_batsman = top_batsman.reset_index()
top_batsman
```

```python
# Lets plot a bar plot of the result2 usign striker and runs_off_bat using matplotlib
plt.figure(figsize=(4,4))
plt.bar(top_batsman['striker'], top_batsman['runs_off_bat'],color='skyblue',edgecolor='black')
plt.title('Most Fours in single match')
plt.xlabel('Batsman')
plt.ylabel("No.of Foures")
plt.xticks(rotation=90);
```

```python
#Most fours hit in a Venue
top_fours = cwc[cwc['runs_off_bat'] == 4]
venue_sixes = top_fours['venue'].value_counts().head(10)
venue_sixes
```

```python
# Lets plot the number of fours hit in a Venue
plt.figure(figsize=(5,4))
plt.bar(venue_sixes.index,venue_sixes,color='skyblue',edgecolor="black")
plt.title("Number of four hit in a venue")
plt.xlabel("Venue Name")
plt.ylabel("No.of foures")
plt.xticks(rotation=90);
```

```python
#Top 10 Teams which scored most number of runs in a match
team=(cwc[cwc['innings'] == 1].groupby(['match_id','batting_team'])['runs_off_bat'].sum() 
      + cwc[cwc['innings'] == 1].groupby(['match_id','batting_team'])['extras'].sum()).sort_values(ascending=False).head(10)
team
```

```python
# Top 10 Batsmen who scored most number of runs in a match
most_run=cwc.groupby(['match_id','striker'])['runs_off_bat'].sum().sort_values(ascending=False).head(10)
most_run
```

```python
# reset the index of result3
most_run = most_run.reset_index()
most_run
```

```python
# Lets plot a bar plot of the result3
plt.figure(figsize=(5,4))
plt.bar(most_run['striker'], most_run['runs_off_bat'],color='skyblue',edgecolor='black')
plt.title('Players who scored most runs in a single match')
plt.xlabel("Batsman")
plt.ylabel("Most Run")
plt.xticks(rotation=90);
```

```python
#Overall Team Score in the season
teamscore=cwc.groupby('batting_team')['runs_off_bat'].sum()
+ cwc.groupby('batting_team')['extras'].sum().sort_values(ascending = False)
teamscore
```

```python
# lets plot the bar chart of the overall team score in the tournament
plt.figure(figsize=(5,4))
plt.bar(teamscore.index,teamscore)
plt.title("Overall Team Score in ICC WCC 2023")
plt.xlabel("Team")
plt.ylabel("Total Runs")
plt.xticks(rotation=90);
```
